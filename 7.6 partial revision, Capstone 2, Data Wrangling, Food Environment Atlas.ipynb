The above graph compares the percent of the population that is both low-income and has low access to stores on the x-axis, with the percent of SNAP households that have low access to stores on the y-axis.  It seems that SNAP benefits are not necessarily a good indicator of poverty.  There doesn't seem to be any clear corrolation between low income and SNAP households among households with low access to food stores.  There are even several counties in which there is a higher percentage of households receiving SNAP benefits than there are households considered to be low-income among households with low access to stores.  If anything, SNAP benefits seem to be more related to what state the household is located in than the income of any household.  The highest percent of the population receiving SNAP benefits can be found in Washington District of Columbia, followed by Mississippi, New Mexico, Oregon, and Tennessee.
