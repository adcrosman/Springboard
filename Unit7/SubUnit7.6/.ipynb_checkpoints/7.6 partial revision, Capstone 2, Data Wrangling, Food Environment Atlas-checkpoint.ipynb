```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
data = pd.ExcelFile('DataDownload.xls')
```

# Goals and Objectives

The goal of this project is to study the distribution of SNAP benefits at country level and look into key drivers for high or low levels of benefit. Specifically we will build of model to predict the SNAP benefits per capita in 2015, given a set of food, store availability and county demographic features. The model will be at County Level.

The model will give us some key insights on any inequality in SNAP benefit distribution across counties. In particular we are interested in finding those Counties for which out model predicts a high level of SNAP benefit per capita, but that instead are underperforming. Those Counties should be the focus for State autorities.

The data is taken from the Food Environment Atlas 2018, published by the United States Department of Agriculture.

```python
print(data.sheet_names)
```

### 1) Explore Assistance Tab

The Assistance Tab provides a set of variables related to the distribution of SNAP benefits across Counties and States. This is the part of the data where we will find our Target Variable, the variable we want to model and predict.

```python
assistance = data.parse('ASSISTANCE')
assistance.head()
```

```python
# This dataset contains many columns; let's see what they mean
assistance_variables = metadata[metadata['Variable Code'].isin(assistance.columns)]
assistance_variables
```

The variable **PC_SNAPBEN15** is our variable of interest. This is the variable that we want to predict given a set of features. 

As we can see below Counties have in average 17.6 dollars per capit, with a Max of 100 dollars per capita in some Counties. The data is not exactly distributed as a Gaussian, but is right skewed. We will make further exploration of this data in future notebooks

```python
target = assistance['PC_SNAPBEN15']
print(target.describe())

# Plot Histogram
plt.figure(figsize = (8,6))
plt.hist(target)
plt.title("Distribution of Snap Benefit Per Capita in 2015 Across different counties")
```

After isolating the target variable we will take a closer look to the features available in the Assistance tab. We want to discard features that relate to period after 2015 and the features that are too far back in the past. We should be very cautious with including Assistance variable for 2015, since those could be too correlated with the target variable and cause Data Leakage https://www.analyticsvidhya.com/blog/2021/07/data-leakage-and-its-effect-on-the-performance-of-an-ml-model/

Therefore interesting features in this part of the data could be.

1. *REDEMP_SNAPS12* = SNAP redemptions/SNAP-authorized stores, 2012
2. *PCT_SNAP12* = SNAP participants (% pop), 2012*
3. *PC_SNAPBEN10* = SNAP participants (% pop), 2010*
4. *SNAP_PART_RATE13* = SNAP participants (% eligible pop), 2013
5. *PCT_FREE_LUNCH14* = Students eligible for free lunch (%), 2014
6. *PCT_REDUCED_LUNCH14* = Students eligible for reduced-price lunch (%),..
7. *PC_WIC_REDEMP12* = WIC redemptions per capita, 2012
8. *REDEMP_WICS12* = WIC redemptions/WIC-authorized stores

The above graph compares the percent of the population that is both low-income and has low access to stores on the x-axis, with the percent of SNAP households that have low access to stores on the y-axis.  It seems that SNAP benefits are not necessarily a good indicator of poverty.  There doesn't seem to be any clear corrolation between low income and SNAP households among households with low access to food stores.  There are even several counties in which there is a higher percentage of households receiving SNAP benefits than there are households considered to be low-income among households with low access to stores.  If anything, SNAP benefits seem to be more related to what state the household is located in than the income of any household.  The highest percent of the population receiving SNAP benefits can be found in Washington District of Columbia, followed by Mississippi, New Mexico, Oregon, and Tennessee.

Even when looking specifically at SNAP households with low access to stores as a percentage of the population, there is no correlation with the number of SNAP authorized stores per thousand people.  This means that may households receiving SNAP benefits may not necessarily have any accessable stores to redeem benefits at.  It is likely that this is because many large chains would not consider it profitable to invest in stores in areas of the United States that are remote and/or have largely low-income customers, and those that choose to open stores in those locations may not consider it worthwile to bother accepting SNAP benefits.  That being said, SNAP benefits themselves are not being consistently distributed to low-income households, so unless and until they are distributed in a way that is more indicative of need, those looking to ensure equitable food distribution in the United States would do well to look beyond the SNAP program.
